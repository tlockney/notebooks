```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 10, 20)
y = x**2
```

```python
plt.plot(x,y)
plt.title('Our first plot')
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
```
